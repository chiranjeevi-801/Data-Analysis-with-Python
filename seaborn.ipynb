# Data Visulization  With Seaborn

### Seaborn is a Python data visualization library built on top of Matplotlib that provides a high-level interface for drawing informative and attractive statistical graphics. It simplifies complex plots, integrates well with Pandas DataFrames, and includes built-in themes and color palettes for cleaner visuals.



#### Key Features of Seaborn:-----

High-level abstraction – easier to use than raw Matplotlib.

Works directly with Pandas DataFrames.

Built-in statistical plots – histograms, bar plots, scatter plots, box plots, violin plots, heatmaps, etc.

Beautiful default styles – better aesthetics out of the box.

Supports categorical and continuous data visualization.

Integration with NumPy, Pandas, SciPy, and Matplotlib.

```python
import seaborn as sns 
```

```python
## Basic Plotting With Seaborn 

```

```python
tips=sns.load_dataset('tips')
tips
```

```python
 # create a scatter plot 
import matplotlib.pyplot as plt 

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter Plot pf Total Bill vs Tip")
plt.show()
```

```python
## Line Plot 

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bil by size")
plt.show()
```

```python
## Categorical Plots 
## bar Plot 

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill By Day')
plt.show()
```

```python
## Box Plot 

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
## Violin Plot 

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
### Histograms 

sns.histplot(tips['total_bill'],bins=10,kde=True) ## kde--> it is curve 
```

```python
sns.histplot(tips['total_bill'],bins=10,kde=False) ## kde--> it is curve 
```

```python
## KDE Plot 

sns.kdeplot(tips['total_bill'],fill=True)
```

```python
## PairPlot 

sns.pairplot(tips)
```

```python
tips 
```

```python
## Heatmap

corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import pandas as pd 
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## Plot total sales by product 

plt.figure(figsize=(10,6))
sns.barplot(x='Product_Category',y='Discount', data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Total sales')
plt.show()
```

```python
## Plot total sales by Unit_Cost	

plt.figure(figsize=(10,6))
sns.barplot(x='Unit_Cost',y='Discount', data=sales_df,estimator=sum)
plt.title('Total Sales by Unit_Cost')
plt.xlabel('Unit_Cost')
plt.ylabel('Total sales')
plt.show()
```
