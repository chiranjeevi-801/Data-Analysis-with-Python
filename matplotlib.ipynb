```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## Create a line plot 

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
## Create  a customized line plot 

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='green',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
## Multiple Plots
## Smpple data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1) ## 1,2 means i want do create how many rows and colomns in this abvoe figure !-row and 2-colomns / 
plt.plot(x,y1,color='green')
plt.title("Plot 1")


plt.subplot(2,2,2) 
plt.plot(y1,x,color='red')
plt.title("Plot 2")


plt.subplot(2,2,3) 
plt.plot(x,y2,color='blue')
plt.title("Plot 3")


plt.subplot(2,2,4) 
plt.plot(x,y2,color='yellow')
plt.title("Plot 4")


```
